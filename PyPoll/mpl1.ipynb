```python
import numpy as np
import matplotlib.pyplot as mpl

```

```python
x_axis = np.arange(1, 13, 1)
x_axis
```

```python
points = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
mpl.plot(x_axis, points)
mpl.show
```

```python
points_C = [(x-32) * 0.56 for x in points]
points_C
```

```python
mpl.plot(x_axis, points_C)
```

```python
mpl.plot(x_axis, points, marker="+", color="blue", label="Fahrenheit")
mpl.plot(x_axis, points_C, marker="s", color="red", label="Celsius") 
mpl.show
```

```python
fahrenheit, = mpl.plot(x_axis, points, marker="+",color="blue", linewidth=1, label="Fahreneit")
celsius, = mpl.plot(x_axis, points_C, marker="s", color="Red", linewidth=1, label="Celsius")

mpl.xlabel("Months")
mpl.ylabel("Degrees")

mpl.legend(handles=[fahrenheit, celsius], loc="best")
```
